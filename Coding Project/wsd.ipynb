```python
def removePunctuation(sentences):
    """
    Removing punctuation
    :param sentences:
    """
    tt=""
    sentences2=[]
    for x in sentences:
        text2=x
        cnt=1
        for x2 in text2:
            if cnt==1:
                tt+=x2
                cnt=0
            else:
                tt+=" "+x2    
        sentences2.append(tt)
        tt=""
    return sentences2
```

```python
def stopWords(sentences2):
    """
    Removing stopwords
    """
    stop_words = set(stopwords.words("english"))
    context_tab=[]
    for sentence in sentences2:
        words = nltk.word_tokenize(sentence)
        without_stop_words = [word for word in words if not word in stop_words]
        context_tab.append(without_stop_words)
    return context_tab 
```

```python
def lemmatization(context_tab):
    """
    Lemmatization
    """
    lemma=[]
    wl=WordNetLemmatizer()
    for x in context_tab:
        m2=[]
        for x2 in x:
            x3=wl.lemmatize(x2,wn.VERB)
            x3=wl.lemmatize(x3,wn.NOUN)
            x3=wl.lemmatize(x3,wn.ADJ)
            x3=wl.lemmatize(x3,wn.ADV)
            m2.append(x3)
        lemma.append(m2)
    return lemma
```

```python
def tagPos(lemma):
    """
    tagPos
    """
    pos=[]
    for n in lemma:
        pos.append(nltk.pos_tag(n))
    return pos   
```

```python
def wordOfInterest(pos):
    """
    Keyword Extraction
    """
    wn_pos=['NN','VB','JJ','JJR','JJS','NNP','VBG','RB','VBD','VBP']

    woi1=[]

    for x in pos:
        arr=[]
        for y in x:
            if y[1] in wn_pos:
                arr.append(y)
        woi1.append(arr) 
    woi=[]


    for i in woi1:
        arr2=[]
        for j in i:

            if j[1]=='VBD' or j[1]=='VB' or j[1]=='VBP':
                tup=(j[0],'v')
                arr2.append(tup)
            elif j[1]=='VBG':
                tup=(j[0],'n')
                arr2.append(tup)
            elif j[1]=='NN' or j[1]=='NNP':
                tup=(j[0],'n')
                arr2.append(tup)
            elif j[1]== 'JJ' or j[1]=='JJR' or j[1]=='JJS':
                tup=(j[0],'a')
                arr2.append(tup)
            elif j[1]=='RB':
                tup=(j[0],'r')
                arr2.append(tup)
        woi.append(arr2)       
            
    return woi
```

```python
def pre_process(sentences):
    """
    Pre-processing of sentences
    """
    semcor_data_without_punctuation = removePunctuation(semcor_data)
    semcor_data_without_stop_words = stopWords(semcor_data_without_punctuation)
    semcor_data_lemmatization = lemmatization(semcor_data_without_stop_words)
    wordPos=tagPos(semcor_data_lemmatization)
    woi=wordOfInterest(wordPos)
    return woi
semcor_data = semcor.sents()
woi = pre_process(semcor_data)
```

```python
def get_sense(word,tag):
    if tag == 'n':
        sense = wn.synsets(word,pos=wn.NOUN)
    elif tag == 'v':
        sense = wn.synsets(word,pos=wn.VERB)
    elif tag == 'a':
        sense = wn.synsets(word,pos=wn.ADJ)
    elif tag == 'r':
        sense = wn.synsets(word,pos=wn.ADV)
    return sense
```

```python
#Statistical word frequency
def word_freq(woi):
    word_frequency = {}
    for sentence in woi:
        for word in sentence:
            if word[0] in word_frequency:
                word_frequency[word[0]] += 1
            else:
                word_frequency[word[0]] = 1
    return word_frequency
word_frequency = word_freq(woi)
word_frequency
```

```python
def get_features(sentence,word,index):
    context = []
    pos_context = []
    for i in range(max(0, index-3), index):
        context.append(sentence[i][0])
        pos_context.append(sentence[i][1])
    for i in range(index, min(index+3, len(sentence))):
        # print("i",i,"index",index,"len(sentence)",len(sentence))
        context.append(sentence[i][0])
        pos_context.append(sentence[i][1])
    feature = {
        'word': word,
        'context': tuple(context),
        'pos_context': tuple(pos_context),
        'pos': sentence[index][1],
        'index': index,
        'word_frequency': word_frequency[word]
    }
    return feature
```

```python
# prepare_data
def prepare_data(word):
    data = []
    for sentence_and_tag in woi:
        for idx,word_and_tag in enumerate(sentence_and_tag):
            if word_and_tag[0].lower() == word.lower():
                senses = get_sense(word_and_tag[0],word_and_tag[1])
                for sense in senses:
                    feature = get_features(sentence_and_tag,word,idx)
                    data.append((feature,sense.name()))
    return data
```
